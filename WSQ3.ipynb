#Day 11: Create a list and calculate the sum of the squares of each of the numbers added

```python
#Use of list comprenhension to sum and square the number in the empty list called numbers
def sumsquares_list(numbers):
    total = sum(num ** 2 for num in numbers)
    return total

def main():
    numbers = []

    print("Hello! I will ask you to give me numbers. Ready?")

    while True:
        try:
            num = float(input("Please enter a number (or type S to stop): "))
            numbers.append(num)
        except ValueError:
            break

    if numbers:
        result = sumsquares_list(numbers)
        print("The sum of the squares of the numbers you provided is:", result)
    else:
        print("No numbers were provided.")

if __name__ == "__main__":
    main()
```

#Day 12: "I am 196" and how to find a lychrel number

```python
#A Lychrel number is a natural number that cannot
#form a palindrome through the iterative process of
#repeatedly reversing its digits and adding the resulting numbers
# E.g. 123 + 321 = 444

def invert(n):
    s = str(n)
    s = s[::-1]
    inversa = int(s)
    return inversa

def main():
    less = int(input("Enter the lower value: \n"))
    high = int(input("Enter the higher value: \n"))

    palindrome = 0
    lychrel = 0
    natural = 0 #natural palindrome like 11,22,33
    lychrel_num = []

    for counter in range(less, high + 1):
        t = 0
        for m in range(counter, counter +1):
            t = invert(m)
            sum_val = t + counter

        if counter == t:
            natural += 1
        else:
            repetition = 0
            candidate = counter
            t = invert(candidate)

            #Test if the number becomes palindrom after nth repetitions
            while candidate != t and repetition < 30:
                candidate = candidate + t
                t = invert(candidate)
                repetition += 1

            if repetition < 30:
                palindrome += 1
            else:
                #print("Lychrel found", counter) to get the count instead of the numbers
                lychrel += 1
                lychrel_num.append(counter)

    print("\nNatural Palindrome:", natural)
    print("Number of Non-lychrel :", palindrome)
    print("Lychrel Candidates:", lychrel_num)

if __name__ == "__main__":
    main()
```

#Day 13: Greatest Common Divisor

```python
#Use of the euclidian algorithm to get the GCD

def common(first, second):
    if first == second:
        return first
    elif first > second:
        return common(first - second, second)
    else:
        return common(first, second - first)

def main():
    print("Hello! This is the greatest common divisor calculator.")

    x = int(input("Enter the first number, please: "))
    y = int(input("Enter the second number, yeah! "))

    gcd = common(x, y)

    print("The greatest common divisor is", gcd)

if __name__ == "__main__":
    main()
```

#Day 14: Babylonian method for square root

```python
def Babylonian(n):
    error = 0.0001 #error decides the accuracy level
    s = n

    while abs(s - n/s) > error:
        s = (s + n/s) / 2
        print("The intermediate number is", s)

    return s

def main():
    n = float(input("Insert a number to get the square root with the Babylonian method\n"))

    a = Babylonian(n)

    print(f"The root in Babylonian method is: {a}")

if __name__ == "__main__":
    main()
```

#Day 15: Image resize or b/w

```python
import matplotlib.pyplot as plt
import numpy as np

from skimage import data #To get the astronaut image
from skimage import color
from skimage.transform import resize


def main():
    try:
        print("What do you want to do? (press the number on your keyboard)")
        print("1. Resize the image to one half of its original size")
        print("2. Switch to black and white")

        image = data.astronaut()
        x = int(input())

        if x == 1:
            resize_photo = resize(image, (image.shape[0] // 2, image.shape[1] // 2))
            plt.subplot(1, 2, 1)
            plt.imshow(image)
            plt.title("Original Image")
            plt.subplot(1, 2, 2)
            plt.imshow(resize_photo)
            plt.title("Resized Image")
            plt.show()

        if x == 2:
            grayscale_photo = color.rgb2gray(image)
            plt.subplot(1, 2, 1)
            plt.imshow(image)
            plt.title("Original Image")
            plt.subplot(1, 2, 2)
            plt.imshow(grayscale_photo, cmap='gray')
            plt.title("Grayscale Image")
            plt.show()

    except Exception as error:
        print("Caught exception:", error)
        return 1

if __name__ == "__main__":
    main()

```
